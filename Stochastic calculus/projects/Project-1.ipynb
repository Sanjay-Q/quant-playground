Your team has been asked to design a regime-based allocation strategy that rotates
among TLT (US Treasury ETF), GLD (Gold ETF), and SPY (S&P 500 ETF) depending on
the market’s volatility environment, measured through VIX.
The objective is to use Markov chains or Hidden Markov Models (HMMs) to infer
volatility regimes and adapt allocations accordingly.

Your goal is to implement and evaluate this strategy in a transparent, step-by-step way
that an instructor can easily follow and assess.
Tasks
Step 1: Data Preparation and Exploration
1. Download daily adjusted close prices for TLT, GLD, SPY, and VIX from Yahoo
Finance, using the maximum common sample period for which all series have
available data.
2. Compute daily log-returns for ETFs and the change or return of VIX.
3. Align the data series on common dates, remove missing values, and visualize:
● One plot showing ETF returns over time.
● One plot showing ∆VIX (or VIX returns).

Deliverable: Clean data and clear exploratory plots.
Objective: Verify data consistency and build intuition on volatility dynamics.



Step 2: Modeling VIX Regimes
Discrete Markov Chain Option:
● Discretize ∆VIX into 2–3 states (e.g., low/medium/high
volatility) using quantiles.
● Estimate the transition matrix and the stationary distribution.
Hidden Markov Model Option:
● Fit 2- and 3-state Gaussian HMMs to ∆VIX using the EM algorithm.
● Extract the most likely state sequence and smoothed probabilities.
Deliverable: Transition matrix (MC) or fitted parameters (HMM) and a plot of VIX with
color-coded states.

Step 3: State Selection and Interpretation
1. Compare candidate models using log-likelihood, AIC and BIC.
2. Select the preferred model (number of states + method) and explain why.
3. Using the chosen state sequence, compute mean and standard deviation of ETF
returns by state, and visualize results with a bar chart.
Deliverable: Table of model comparison, chosen model, and bar chart of
state-conditional returns.
Objective: Choose an interpretable and stable regime model and relate states to asset
behavior.

Step 4: Designing the Rotation Strategy
1. Define a simple, rule-based allocation using the identified regimes:
○ For each state, allocate 100% to the ETF with the highest historical mean
return in that state.
○ Optionally, use 60/40 weights between the top two ETFs or introduce a
1-day execution lag to avoid lookahead bias.

2. Summarize the final state → allocation mapping in a small table.
Deliverable: Clear decision rule (state → asset allocation) and brief
justification.
Objective: Translate statistical regimes into a practical investment policy.


Step 5: Backtesting and Evaluation
1. Backtest the strategy over the sample period using a 1-day execution lag.
2. Compute performance metrics:
○ Cumulative Return, Annualized Return, Volatility, Sharpe Ratio, and Max
Drawdown.
3. Compare results to two benchmarks:
○ Equal-weight (1/3 each, monthly rebalanced)
○ Buy-and-hold SPY
4. Plot cumulative performance curves and summarize results in a table.
5. Briefly discuss sensitivity to number of states or sample window.
Deliverable: Performance plots and summary table with short interpretation.
Objective: Evaluate whether regime-based rotation improves risk-adjusted performance
relative to simple benchmarks.
